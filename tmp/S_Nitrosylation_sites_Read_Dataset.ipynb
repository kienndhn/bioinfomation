# Dataset: 
Tập dữ liệu training: 3383 positive and 3365 negative sites (hiện tại đã có)

Tập dữ liệu test: contain 351 positive and 3168 negative sites (hiện tại chưa có, mới chỉ có vị trí ở trong file S-nitrosylation datasets-PreSNO_test.csv )



```python
file_test_index = "S-nitrosylation datasets-PreSNO_test.csv"
file_protein_sequences = "protein_30_phan_tram_cutoff.csv"
```

```python
f = open(file_protein_sequences, "r")
protein_sequences = f.readlines()
f.close()

print("Tổng số protein: ", len(protein_sequences))
print(protein_sequences[0])

# Lưu các sequences vào một dictionary: key là id của protein 
sequences = {}
for line in protein_sequences:
    seq = line.strip().split(",")[1]
    #print(seq)
    seq_id = line.strip().split(",")[0].split("|")[1]
    #print(seq_id)
    
    sequences[seq_id] = seq
    #break

print("Tổng số sequences đã xử lý: ", len(sequences))
```

```python
# Đếm tổng số ký tự C trong các chuỗi protein
total_count = 0
for line in protein_sequences:
    seq = line.strip().split(",")[1]
    total_count += seq.count("C")
print("Total count C: ", total_count)
```

```python
f = open(file_test_index, "r")
lst_index = f.readlines()
f.close()

print("Tổng số mẫu trong tập Test: ", len(lst_index) - 2)
print(lst_index[0])
print(lst_index[1])
print(lst_index[2])

print("#### Check từng vị trí xem có đúng là ký tự C trong protein sequence? ")
nb_correct = 0
nb_correct_positive = 0
nb_correct_negative = 0
for i, line in enumerate(lst_index[2: ]):
    #print(line)
    seq_id = line.strip().split(",")[0].split("|")[1]
    #print("Sequence id: ", seq_id)
    C_index = int(line.strip().split(",")[1])
    #print("C index: ", C_index)
    label = line.strip().split(",")[2]
    #print("Label: ", label)
    
    #print(sequences[seq_id][C_index-1])
    if seq_id not in sequences.keys(): 
        print("ERROR : sequence id not in list => ", seq_id)
    elif C_index > len(sequences[seq_id]):
        print("ERROR: ", i, C_index, len(sequences[seq_id]))        
    elif (sequences[seq_id][C_index-1] != 'C'):
        print("ERROR", sequences[seq_id][C_index-1])
    else: 
        nb_correct += 1
        if label == "SNO": nb_correct_positive += 1
        if label == "Non-SNO": nb_correct_negative += 1            
    
    #break
    
print("Tổng số mẫu đã check đúng ký tự C: ", nb_correct)
print("Tổng số mẫu đã check đúng ký tự C (positive): ", nb_correct_positive)
print("Tổng số mẫu đã check đúng ký tự C (negative): ", nb_correct_negative)
```
